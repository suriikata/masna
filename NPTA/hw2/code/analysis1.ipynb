The first dataset comprises of the most famous Ski Resort in the entire world, with 33 features. In order to simplify the process of analysis, I initially extracted 10 features, that will be suitable for further non-parametrically oriented analysis.

```r
library(readr)
library(dplyr)
library(skimr)
```

```r
ski_data <- read.csv("data/resortworldwide.csv", header = TRUE)
head(ski_data)
names(ski_data)
```

```r
ski_data_selected <- ski_data %>%
                select(NameResort, Continent, Country, Rate, Altitude, Easy, Intermediate, Difficult, Adult, Total.Kms)


ski_data_selected$NameResort <- as.factor(ski_data_selected$NameResort)
ski_data_selected$Continent <- as.factor(ski_data_selected$Continent)
ski_data_selected$Country <- as.factor(ski_data_selected$Country)

ski_data_selected$Adult <- as.numeric(gsub(",", ".", ski_data_selected$Adult))

head(ski_data_selected)
skim(ski_data_selected)
```

The Rate variable contains an excessive number of missing values, which surpasses the acceptable threshold of around 10% missing values. As a result, the missing values representing around 35% of the intial sample were excluded, but leaving a sufficient sample size. Additionally, there were missing values in the Altitude and Adult variables, but the proportion of the latter is negligible (les than 1%) and also excluded, ultimately resulting in 3466 units.  

```r
ski_smaller <- ski_data_selected[complete.cases(ski_data_selected[, c("Rate", "Altitude", "Adult")]), ]

ski_smaller$NameResort <- as.factor(ski_smaller$NameResort)
ski_smaller$Continent <- as.factor(ski_smaller$Continent)
ski_smaller$Country <- as.factor(ski_smaller$Country)

skim(ski_smaller)
```

```r
# normality distribution of numeric variables

ski_smaller_numeric <- select(ski_smaller, 4:10)
head(ski_smaller_numeric)

options(repr.plot.width = 20, repr.plot.height = 10)
par(mfrow = c(1,4), cex.main = 3, cex.axis = 1.5)
for (p in 1:7){
    hist(ski_smaller_numeric[,p], main = paste(names(ski_smaller_numeric)[p], " | Initial Hist"), col = p)
    hist(ski_smaller_numeric[,p], main = paste(names(ski_smaller_numeric)[p], " | F-D Optimiz."), col = p, breaks = "Freedman-Diaconis")
    boxplot(ski_smaller_numeric[,p], main = names(ski_smaller_numeric)[p], col = p)
    plot(density(ski_smaller_numeric[,p]), main = names(ski_smaller_numeric)[p], col = p)
}

dev.off()
```

The analysis reveals that all variables exhibit a strong right skewness, with rate and altitude having a relatively even distribution of values on the right hand side of the histograms and density plots. However, the remaining variables demonstrate extreme values, which result in a prominent long tail in histogram as well as in density plot. In order to obtain meaningful results, it will be necessary to exclude these extreme values from these five variables to achieve effective clustering.

```r
# statistical testing of normality

library(nortest)


# For Shapiro-Wilk
ski1 <- data.frame("Variable" = NA,
                  "Test Stat" = NA,
                  "PV" = NA,
                  "Method" = NA)

# For Anderson-Darling
ski2 <- data.frame("Variable" = NA,
                  "Test Stat" = NA,
                  "PV" = NA,
                  "Method" = NA)

options(scipen = 0) # return back the notations
for (t in 1:7){
    ski1[t,] <- data.frame(
        "Variable" = names(ski_smaller_numeric)[t],
        "Test Stat" = shapiro.test(ski_smaller_numeric[,t])$statistic,
        "PV" = shapiro.test(ski_smaller_numeric[,t])$p.value,
        "Method" = shapiro.test(ski_smaller_numeric[,t])$method
    )
}

for (t in 1:7){
    ski2[t,] <- data.frame(
        "Variable" = names(ski_smaller_numeric)[t],
        "Test Stat" = ad.test(ski_smaller_numeric[,t])$statistic,
        "PV" = ad.test(ski_smaller_numeric[,t])$p.value,
        "Method" = ad.test(ski_smaller_numeric[,t])$method
    )
}

rbind(ski1, ski2)
```

The Shapiro-Wilk and Anderson-Darling normality tests show statistically significant p-values (i.e., p < 0.05) for all numerical variables, meaning that the null hypothesis of normality is rejected for each variable. In other words, there is evidence to suggest that the data for each variable do not follow a normal distribution and the previous assumption made based on visual graphs are confirmed. 

```r
library(corrplot)
ski_num_cor <- cor(as.matrix(ski_smaller_numeric), method = "kendall")
ski_num_cpv <- cor.mtest(ski_num_cor, conf.level = 0.95)
corrplot(ski_num_cor, p.mat = ski_num_cpv$p, method = "square", col = terrain.colors(10), type = "full", hclust.method = "median")
```

The correlation analysis using corrplot revealed that the majority of the numerical variables were not correlated, except for a strong association between the rate of the ski resort and the number of easy and intermediate tracks, prompting me to select these three variables for further analysis.

### Density Functional Analysis / Clustering

When considering cluster algorithm setups for the first dataset, which is characterized by continuous variables, non-normal distribution, and the presence of significant outliers, the most promising method appears to be median-based clustering. This is due to the fact that this method uses the median to determine the cluster centers, which is less sensitive to extreme values than the mean. Another viable clustering option for the first dataset is density estimation via kernel density estimators, as this method can effectively handle noise and outliers, typically excluding them from the clusters due to their low density estimates. Conversely, binary MCMC coupled-Metropolis clustering is not a suitable choice, given that our variables are predominantly continuous. If necessary and hypothethically, it could be possible to convert the variables into dummy variables or apply a threshold value to assign 0 or 1 to each variable.

```r
# median based
library(Gmedian)
library(RColorBrewer)

set.seed(157)

ski_gm <- kGmedian(X = ski_smaller_ei[, 4:10], ncenter = 6, gamma = 1, alpha = 0.85, nstart = 25, iter = 999)
str(ski_gm)

barplot(table(ski_gm$cluster), col = brewer.pal(6, name = 'Set3'), 
        main = 'Median-based Clustering', ylim = c(0, 700)) 
```

```r
with(ski_smaller_ei[,4:10],
    boxplot(Rate ~ ski_gm$cluster, col = brewer.pal(6, name = 'Set3'), xlab = 'Number of Clusters'))
```

```r
with(ski_smaller_ei[,4:10],
    boxplot(Easy ~ ski_gm$cluster, col = brewer.pal(6, name = 'Set3'), xlab = 'Number of Clusters'))
```

```r
with(ski_smaller_ei[,4:10],
    boxplot(Intermediate ~ ski_gm$cluster, col = brewer.pal(6, name = 'Set3'), , xlab = 'Number of Clusters'))
```

During my analysis using gmedian function, I observed that reducing the number of clusters resulted in less differentiation between the clusters based on the selected 3 variables, but the advantage of this was observed in that the groups have become comparable in size.

Ultimately, I decided to use 6 clusters, because the differences between clusters were apparent and varied between the clusters, and secondly because I assumed that the clustering would be capable of separating ski resorts by continent, of which there were 6. Hence, before interpreting the gmedian output I decided to explore the distribution of these three variables across continents using boxplot. A trend was be observed that ski resorts with a higher rating typically have more trails with easy and medium difficulty, particularly Ocenia, South and North America, as well as Asia. Looking at the clusters obtained, results indicate that the two smaller clusters have a high average rating (e.g. 2 and 4) and a wider range of easy and challenging routes compared to those with a lower rating.  I assume that this clusters contain American or Asian ski resorts, based on prior exploratory analysis. ### Regression Analysis

```r
# kernel algorithm
library(gplm) 

Rate_ka  <- kde(ski_smaller_ei$Rate)
Rate_ka2 <- kde(ski_smaller_ei$Rate, bandwidth=3, kernel="epanechnikov")
Rate_ka3 <- kde(ski_smaller_ei$Rate, bandwidth=3, kernel="triangle")
Rate_ka4 <- kde(ski_smaller_ei$Rate, bandwidth=3, kernel="biweight")
par(mfrow = c(2,2))
plot(Rate_ka)
plot(Rate_ka2, main = "Epanechnikov kernel opt.")
plot(Rate_ka3, main = "Triangle kernel opt.")
plot(Rate_ka4, main = "biweight kernel opt.")
```

During the initial optimization process using kernel algorithm, the Epanechnikov, triangle, and biweight optimization processes were evaluated and compared. The results indicated that the triangle kernel optimization process provided the most promising results. This was likely due to the fact that the triangle kernel function balanced the trade-off between robustness and smoothness in the estimated density, making it more effective in capturing important features in the data, for instance the peak, while also being less sensitive to outliers.

```r
# median based lm
library(mblm)

model1 <- mblm(Rate ~ Easy, data = ski_smaller_ei)
```

```r
summary(model1)

plot(Rate ~ Easy, data = ski_smaller_ei)
abline(model1, lwd = 3, col = "red")
```

```r
model1.2 <- mblm(Rate ~ Intermediate, data = ski_smaller_ei)
```

```r
summary(model1.2)

plot(Rate ~ Intermediate, data = ski_smaller_ei)
abline(model1.2, lwd = 3, col = "red")
```

Upon analyzing the results of the median-based linear model, it is evident that the variable "easy" is statistically significant at a level of less than 1%. This indicates that a one-kilometer increase in easier tracks is associated with an increase of 0.25 units in ratings. Additionally, the variable "intermediate" also appears to be significant at a 1% level, implying that a one-kilometer increase in intermediate tracks is linked with an increase of 0.27 units in ratings. In both scenarios, we observe a positive and consistent correlation between the length of the track and the ratings of ski resorts. This is evidenced by the positively sloping lines on the graphs. However, the results also reveal a high residual standard error of approximately 0.29 in both cases. This indicates that there is a significant amount of variation in the ratings variable that is not accounted for by the track length. This is demonstrated by a 'cloud' of values around a rating value of 3 on both plots.

```r
# rank-based
library(Rfit)

model2 <- rfit(Rate ~ Easy + Intermediate, data = ski_smaller_ei)
```

```r
summary(model2)
```

```r
# GAM modelling
library(mgcv)

model3 <- gam(Rate ~ Easy + Intermediate, data = ski_smaller_ei)

summary(model3)
```

```r
gam.check(model3)
plot(ggeffects::ggpredict(model3), facets = TRUE)
```

```r
model4 <- gam(Rate ~ s(Easy) + s(Intermediate), data = ski_smaller_ei)

summary(model4)
plot(model4)
```

Finally, I conducted GAM modeling with and without smoothing parameters. In both instances, the coefficient estimates for "easy" and "intermediate" were statistically significant at a level of 1% and positively associated with ski resort ratings. In the first model, which did not incorporate smoothing, a one-kilometer increase in "easy" and "intermediate" tracks was linked to an increase of 0.12 and 0.15 units in ratings, respectively and holding the rest of the factors contant.

```r
print(anova(model3, model4, test = "Chisq"))
```

Although the R-squared and GCV values remained unchanged even after incorporating smoothing, the graphical representation of the results suggests that smoother values perform better. Despite this, when the chi-square of differences between the models was calculated, it was found that there was no significant difference between normal and smoothed values. That said, in my opinion the GAM (with smothing) appears to provide the most promising results, when compared with the rest of the models.
